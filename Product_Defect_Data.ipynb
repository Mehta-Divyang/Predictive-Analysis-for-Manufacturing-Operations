```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('Manufacturing Dataset.csv') #'Manufacturing Dataset.csv' file downloaded from https://www.kaggle.com/datasets/nair26/predictive-maintenance-of-machines?resource=download
```

```python
dataframe
```

```python
dataframe = dataframe[['Product ID', 'Process temperature [K]', 'Operational Hours', 'Failure Type']]
dataframe
```

```python
dataframe.isnull().sum(axis=0)
```

```python
dataframe = dataframe.rename(columns={'Product ID':'Machine_ID','Process temperature [K]':'Temperature','Operational Hours':'Run_Time','Failure Type':'Downtime_Flag'})
```

```python
dataframe['Downtime_Flag'] = np.where((dataframe['Downtime_Flag']=='No Failure'),0,1)
```

```python
dataframe
```

```python
dataframe.to_csv("machine_maintenance_data.csv",index=False)
```
